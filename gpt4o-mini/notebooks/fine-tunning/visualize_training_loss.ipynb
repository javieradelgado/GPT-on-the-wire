```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("TrainingLoss/ICMPv1.csv", sep="\t").set_index('Step')
# reverse the order of rows
df = df.iloc[::-1]
df
```

```python
TL="Final Training Loss = " + str(df["Training loss"][280])

plt.figure(figsize=(10,5))
plt.title("GPT-3.5 turbo-Instruct Training Loss During ICMP Fine-tuning")
plt.plot(df['Training loss'].to_numpy(),label=TL)

plt.xlabel("Steps")
plt.ylabel("Loss")
plt.legend()
plt.savefig("visualizations/ICMP/35TurboInstruct_loss.png",dpi=300)
plt.show()
```
